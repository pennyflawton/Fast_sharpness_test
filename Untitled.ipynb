```python
import numpy as np
from matplotlib import pyplot as py
import skimage.filters
from PIL import Image
import imagesFromMovie
from fastmlv import fastmlv_image
```

```python
images = imagesFromMovie.capture('data_stream_separated_001raw_movie.mov')
imdata = images[0,:,:,0]


sigma = 2.0

blurred = skimage.filters.gaussian(imdata, sigma = (sigma, sigma), truncate = 3.5)

fig_1 = py.figure(1, figsize=(8,8))
ax1_1 = fig_1.add_subplot(121)
ax2_1 = fig_1.add_subplot(122)
ax1_1.imshow(imdata,vmin=0,vmax=255)
ax2_1.imshow(blurred,vmin=0,vmax=255)
```

```python
def mlv(data,x,y):
    a=np.array((data[x+1,y],data[x-1,y],data[x,y+1],data[x,y-1],data[x+1,y+1], data[x-1,y-1],data[x+1,y-1],data[x-1,y+1]))
    b=data[x,y]*np.ones((8))
    c=abs(b-a)
    d=max(c)
    return d
    
def mlv_image(imagedata):
    length = imagedata.shape[0]
    width = imagedata.shape[1]
    
    mlvmap = np.zeros((length,width))
    for idx,x in np.ndenumerate(imagedata):
            mlvmap[idx]=mlv(imagedata,idx[0]-1,idx[1]-1)
            
    mlvmap=mlvmap[1:length-1,1:width-1]
    
            
    return mlvmap
          
```

```python
%timeit mlv_image(imdata)
%timeit mlv_image(blurred)
```

```python
%timeit fastmlv_image(imdata)
```

```python
fig_2 = py.figure(2, figsize=(8,8))
ax1_2 = fig_2.add_subplot(121)
ax2_2 = fig_2.add_subplot(122)

sharpmlv=mlv_image(imdata)
blurmlv=mlv_image(blurred)

ax1_2.imshow(sharpmlv,vmin=0,vmax=255)
ax2_2.imshow(blurmlv,vmin=0,vmax=255)
```

```python
bins=255
hist1,edges=np.histogram(sharpmlv,bins)
hist2,edges2=np.histogram(blurmlv,bins)

fig_3 = py.figure(3, figsize=(8,4))
ax1_3 = fig_3.add_subplot(121)
ax2_3 = fig_3.add_subplot(122)

ax1_3.plot(edges[1:],hist1)
ax1_3.set_xlim(0,255)
ax2_3.plot(edges2[1:],hist2)
ax2_3.set_xlim(0,255)
```
